```python
from matplotlib import collections as matcoll
Npoints = 20
x = np.arange(0,Npoints)
y = np.random.normal(loc=10, scale=2, size=Npoints )
lines = []
for i in range(Npoints):
    pair=[(x[i],0), (x[i], y[i])]
    lines.append(pair)
linecoll = matcoll.LineCollection(lines)
fig, ax = plt.subplots()
ax.add_collection(linecoll)

plt.scatter(x,y, marker='o', color='blue')
plt.xticks(x)
plt.ylim(0,40)
plt.show()
ylim=(0,10)
```

# Looking for Models of the Data

```python
# curve made up of sin waves 
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,10,0.1)
y = np.sin(x) + np.sin(2*x) + np.sin(4*x)
plt.plot(x,y)
```

```python
# plot each of the sine waves in y
plt.plot(x,y)
plt.plot(x,np.sin(x))
plt.plot(x,np.sin(2*x))
plt.plot(x,np.sin(4*x))
plt.show()
```

```python
# given x, y can we find the fourier transform of y?
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
# we know x and y
x = np.arange(0,10,0.1)
y = np.sin(x) + np.sin(2*x) + np.sin(4*x)
# find the fourier transform of y
coef = fft.fft(y)

# plot the first 20 coefficients
plt.plot(coef[0:10])



```
