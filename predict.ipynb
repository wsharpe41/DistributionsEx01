```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data00.csv in dataframe
df = pd.read_csv('data00.csv')
# plot the data
#df.plot()
# test the data for stationarity
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['data'])
print('p-value: {}'.format(result[1]))

# plot the autocorrelation
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
#plot_acf(df['data'])
# plot the first difference
df['data'].diff().plot()

#plt.show()
```

```python
# plot the first difference of the data difference
df['data'].diff().diff().plot()
# Get the mean of the second difference
df['data'].diff().diff().mean()
# take the moving average of the data second difference
df['data'].diff().diff().rolling(20).mean().plot()
```
